# Subjectivity scoring

```python
%cd ../../
from subjectivity.subjectivity_classifier import SubjectivityClassifier
classifier = SubjectivityClassifier(model_filename='./data/save/subj-29.tf',
                                    word_filename='./data/word_embeddings/glove.6B.50d.txt')
```

```python
%cd ../../../../
```

```python
import pandas as pd
import matplotlib.pyplot as plt

manifestos = pd.read_csv('manifesto-forewords/manifestos.csv')
manifestos.year = pd.to_datetime(manifestos.year)
manifestos = manifestos[manifestos.year>='1930-01-01']

sub_pct_lst = list()
obj_pct_lst = list()

for idx,foreword in enumerate(manifestos.foreword):
    res = classifier.classify_sentences_in_text(foreword)
    sub = len(res['subjective'])
    obj = len(res['objective'])
    sub_pct = 100*sub/(sub+obj) 
    obj_pct = 100*obj/(sub+obj)
    
    sub_pct_lst.append(sub_pct)
    obj_pct_lst.append(obj_pct)
    
manifestos['sub_pct'] = sub_pct_lst
manifestos['obj_pct'] = obj_pct_lst
manifestos['net_sub_pct'] = manifestos.sub_pct-manifestos.obj_pct

col_dict = {'LAB':'red','CON':'blue','LIB':'gold','SNP':'yellow','UKIP':'purple','BRX':'purple','NAN':'grey'}

fig,ax = plt.subplots(1, figsize=(12,8))
for party in col_dict.keys():
    party_df = manifestos[manifestos.party==party]
    ax.plot(party_df.year, party_df.net_sub_pct, color=col_dict[party])
ax.set_xlabel("Year"),
ax.set_ylabel("Net subjectivity (%)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

conferences = pd.read_csv('conference-speeches/conference.csv')
conferences.year = conferences.year.apply(lambda year: pd.to_datetime(str(year)+'-01-01'))
conferences = conferences[conferences.year>='1930-01-01']

sub_pct_lst = list()
obj_pct_lst = list()

for idx,speech in enumerate(conferences.content):
    res = classifier.classify_sentences_in_text(speech)
    sub = len(res['subjective'])
    obj = len(res['objective'])
    sub_pct = 100*sub/(sub+obj) 
    obj_pct = 100*obj/(sub+obj)
    
    sub_pct_lst.append(sub_pct)
    obj_pct_lst.append(obj_pct)
    
conferences['sub_pct'] = sub_pct_lst
conferences['obj_pct'] = obj_pct_lst
conferences['net_sub_pct'] = conferences.sub_pct-conferences.obj_pct

col_dict = {'Labour':'red','Conservative':'blue','Liberal':'gold','SDP-Liberal Alliance':'gold','Liberal Democrat':'gold'}

fig,ax = plt.subplots(1, figsize=(12,8))
for idx,row in conferences.iterrows():
    ax.scatter(row.year, row.net_sub_pct, color=col_dict[row.party])
ax.set_xlabel("Year"),
ax.set_ylabel("Net subjectivity (%)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from tqdm import tqdm

current_day = datetime.strftime(datetime.now(), format="%Y-%m-%d")
pm_list = [['Winston Churchill (1)',('1940-05-10','1945-07-26')],
               ['Clement Attlee',('1945-07-27','1951-10-26')],
               ['Winston Churchill (2)',('1951-10-27','1955-04-05')],
               ['Anthony Eden',('1955-04-06','1957-01-09')],
               ['Harold Macmillan',('1957-01-10','1963-10-18')],
               ['Alec Douglas-Home',('1963-10-19','1964-10-16')],
               ['Harold Wilson (1)',('1964-10-17','1970-06-19')],
               ['Edward Heath',('1970-06-20','1974-03-04')],
               ['Harold Wilson (2)',('1974-03-05','1976-04-05')],
               ['James Callaghan',('1976-04-06','1979-05-04')],               
               ['Margaret Thatcher',('1979-05-05','1990-11-28')],
               ['John Major',('1990-11-29','1997-05-02')],
               ['Tony Blair',('1997-05-03','2007-06-27')],
               ['Gordon Brown',('2007-06-28','2010-05-11')],
               ['David Cameron',('2010-05-12','2016-07-13')],
               ['Theresa May',('2016-07-14','2019-07-24')],
               ['Boris Johnson',('2019-07-25',current_day)]]

pmqs = pd.read_csv('hansard-pmqs/hansard_pmqs.csv')
pmqs.date = pd.to_datetime(pmqs.date)
pmqs = pmqs[pmqs.date>='1930-01-01']
pmqs = pd.DataFrame({'date':pd.to_datetime(pmqs.groupby(by='date').answer_text.aggregate(lambda x: ''.join(x)).index),
                     'answer_text':pmqs.groupby(by='date').answer_text.aggregate(lambda x: ''.join(x)),
                     'answerer_party':pmqs.groupby(by='date').answerer_party.aggregate(lambda x: x.mode()),
                     'answerer_name':pmqs.groupby(by='date').answerer_name.aggregate(lambda x: x.mode())})
pmqs = pmqs[pmqs.answerer_party.apply(lambda party: party in ['Conservative','Labour'])]
pmqs['answerer_name'] = pmqs.date.apply(lambda date: [pm for pm,(start_date,end_date) in pm_list if date>=pd.to_datetime(start_date) and date<=pd.to_datetime(end_date)][0])

sub_pct_lst = list()
obj_pct_lst = list()

for idx,answer in tqdm(enumerate(pmqs.answer_text.to_list())):
    res = classifier.classify_sentences_in_text(answer)
    sub = len(res['subjective'])
    obj = len(res['objective'])
    sub_pct = 100*sub/(sub+obj) 
    obj_pct = 100*obj/(sub+obj)
    
    sub_pct_lst.append(sub_pct)
    obj_pct_lst.append(obj_pct)
    
pmqs['sub_pct'] = sub_pct_lst
pmqs['obj_pct'] = obj_pct_lst
pmqs['net_sub_pct'] = pmqs.sub_pct-pmqs.obj_pct

# col_dict = {'Labour':'red','Conservative':'blue'}

# fig,ax = plt.subplots(1, figsize=(12,8))
# for idx,row in tqdm(pmqs.iterrows()):
#     ax.scatter(row.date, row.net_sub_pct, color=col_dict[row.answerer_party])
# ax.set_xlabel("Year"),
# ax.set_ylabel("Net subjectivity (%)")
# plt.show()
```
